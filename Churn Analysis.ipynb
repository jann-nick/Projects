```r
fit = glm(Churn~.,family="binomial",data=churn)
```

```r
nrow(churn)
```

```r
# Identify number of rows on 75% data

smp_size = floor(0.75*nrow(churn))
```

```r
smp_size
```

```r
set.seed(123)
```

```r
# Split data into 75:25 on training:test based on number of rows

train_ind = sample(seq_len(nrow(churn)),size=smp_size)
```

```r
# Set 75% rows to training

train = churn[train_ind,]
```

```r
train
```

```r
# Set remaining 25% to test

test = churn[-train_ind,]
```

```r
test
```

```r
# Fit logistical model on training data

fit = glm(Churn~.,family="binomial",data=train)
```

```r
# Predict test based on fit on train

predictedY = predict(fit, test, type = "response")
```
